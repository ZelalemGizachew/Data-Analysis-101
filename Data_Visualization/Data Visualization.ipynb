```python
import pandas as pd
import matplotlib.pyplot as plt # For Saving the plot as image
```

```python
df_population_row = pd.read_csv('../All_Excel_Data/population_total.csv')
```

```python
df_population_row
```

```python
# Dropping null Values

df_population_row.dropna(inplace=True) # Droppping null values
```

```python
df_population_row
```

```python
df_pivot = df_population_row.pivot(index='year', columns='country', values='population')
```

```python
df_pivot = df_pivot[['Ethiopia', 'Egypt', 'Kenya', 'Sudan', 'India', 'China', 'Russia', 'United States']]
```

```python
df_pivot
```

# 1 Lineplot

```python
df_pivot.plot(kind='line', 
              xlabel='Year', ylabel="Population", 
              title='Population Growth (1955 - 2020)', 
              figsize=(15,8))

# Save plot as png
plt.savefig('lineplot.png')

# Export the pivot table as excel file

df_pivot.to_excel('pivot_table_of_lineplot.xlsx')

# Show plot
plt.show()
```

# 2 Barplot

```python
df_pivot
```

```python
df_pivot_bar = df_pivot[df_pivot.index.isin([2020])]
```

```python
df_pivot_bar = df_pivot_bar.T
```

```python
df_pivot_bar.plot(kind='bar', color='orange', 
                  xlabel='Year', ylabel='Population', 
                  title='Population in 2020', 
                  figsize=(13, 6))


# Save plot as png
plt.savefig('barplot.png')

# Export the pivot table as excel file
df_pivot.to_excel('pivot_table_of_barplot.xlsx')

# Show plot
plt.show()
```

## 2.1 Bar Plot for Group of Data

```python
df_pivot_bar_group = df_pivot[df_pivot.index.isin([1960, 1970, 1980, 1990, 2000, 2010, 2020])]
```

```python
df_pivot_bar_group = df_pivot_bar_group.T
```

```python
df_pivot_bar_group.plot(kind='bar', xlabel='Year', ylabel='Population', 
                        title='Population Growth with the Range of 10 Years (1960 - 2020)', 
                        figsize=(15,8))

# Save plot as png
plt.savefig('group_barplot.png')

# Export the pivot table as excel file
df_pivot.to_excel('pivot_table_of_group_barplot.xlsx')

# Show plot
plt.show()
```
