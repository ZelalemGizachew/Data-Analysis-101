## 1.2 Adding a new column with an array
The Data Frame has 1000 rows so we need to create an array with 1000 elements

```python
# import numpy
import numpy as np
```

```python
language_score = np.arange(0, 1000)
language_score
```

```python
len(language_score)
```

```python
df_exams['language_score'] = language_score
df_exams
```

```python
# Create random integer numbers between 1 and 100
int_language_score = np.random.randint(1, 100, size=1000)
len(int_language_score) # 1000
min(int_language_score) # 1
max(int_language_score) # 99
```

```python
# Adding the new random array value to our Data Frame
df_exams['language score'] = int_language_score
df_exams
```

# 1. Math Operations

## 1.1 Operations in columns

```python
# Test Way
x = 0
for data in df_exams['math score']:
    x += data
print(x)
```

```python
# Easy Way
df_exams['math score'].sum()
```

```python
# count, mean, std, max, and min
df_exams['math score'].count()
df_exams['math score'].mean() # total sum / length of rows
df_exams['math score'].std()
df_exams['math score'].max()
df_exams['math score'].min()
```

```python
# Easier calculation with .describe() method
df_exams.describe()
```

## 1.2. Operations in rows


```python
# Sum of 3 rows
df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']
```

```python
# Average of 3 rows and assigning it to new column
df_exams['average'] = (df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']) / 3
df_exams['average'].round(2)
```
